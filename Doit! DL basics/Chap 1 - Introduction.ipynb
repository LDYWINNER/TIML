# 인공지능

- 인공지능: '사람의 지능을 만들기 위한 시스템이나 프로그램'
	- 예를 들어, 알파고는 '사람처럼 바둑을 둘 수 있게 만들어진 프로그램'
    - 음성 비서는 '사람이 음성으로 질문한 것에 대한 답을 할 수 있도록 만들어진 프로그램'

### 강 인공지능 vs. 약 인공지능
- 강 인공지능: '사람과 구분이 안 될 정도로 강한 성능을 가진 인공지능'
	- 강 인공지능을 만들 수 있는 방법을 아는 사람은 아직 아무도 없으며, 강 인공지능이 나올 시기도 예측하기 어려움
    - 지금까지 발전을 거듭하고 있는 인공지능 기술은 모두 약 인공지능임
- 약 인공지능: 특정 영역에서 작업을 수행하는 인공지능
	- 예: 운전 보조, 질문 답변, 검색, ...
    - 약 인공지능은 특정 영역에서 아주 뛰어난 성능을 발휘하고 있으며 우리의 생활을 보조하는 기술로 많이 사용되고 있음
    - 머신러닝, 딥러닝 모두 약 인공지능에 포함됨

### 머신러닝, 딥러닝, 인공지능의 관계 

딥러닝 $\subset$ 머신러닝 $\subset$ 인공지능

- 강 인공지능은 실제로 만들기 어려운 기술이므로 여기서 말하는 인공지능은 약 인공지능을 지칭함
- 머신러닝을 알고리즘의 종류에 따라 더 세부적으로 나눌 수 있음
- 그 중에 인공신경망 알고리즘을 사용하여 만든 것이 딥러닝임
- 또한 인공신경망의 기본 구성 요소는 다른 머신러닝 알고리즘이 발전된 것임
- 따라서, 딥러닝을 잘 이해하기 위해서는 먼저 머신러닝에 대해 알아야 함

# 머신러닝

- 한글로 풀어 쓰면 '기계 학습'

### '학습'

- 머신러닝과 딥러닝에서 말하는 '학습'은 데이터의 규칙을 컴퓨터 스스로 찾아내는 것을 말함
- 전통적인 프로그램은 사람이 정한 규칙대로 동작하고, 사람이 프로그램의 실행 결과를 보며 규칙을 조금씩 수정함
- 머신러닝은 사람이 만든 프로그램이기는 하지만 규칙을 스스로 찾아 수정함
- 여기서 말하는 규칙을 찾아 수정하는 과정을 '학습' 또는 '훈련'이라고 부름

### 학습 방식

- 머신러닝은 학습 방식에 따라 지도 학습(supervised learning), 비지도 학습(unsupervised learning), 강화 학습(reinforcement learning)으로 분류함
- 머신러닝으로 다루는 많은 작업들이 지도학습에 속함

### 지도학습

- 지도학습은 입력과 타깃으로 모델을 훈련시킴
- 만약 전날의 습도에 따라 다음 날에 비가 왔는지 안 왔는지 꾸준히 기록했다면 오늘의 습도만 알아도 그 다음날 비가 올지 안 올지 예측할 수 있을 것임
- 위의 예시는 지도학습에 속함
- 위와 같은 지도학습을 한 모델은 훈련을 통해 '오늘 습도가 0.6 이상이면 내일 비가 온다'와 같은 '규칙'을 얻을 수 있음
<br><br>
- 훈련 데이터: 모델을 훈련시키기 위해 사용하는 데이터 통틀어서 - 입력과 타깃으로 구성되어 있음
- 입력: 모델이 풀어야 할 일종의 문제와 같은 것
- 타깃: 모델이 맞춰야 할 정답과 같은 것
- 문제에 대한 답을 주는 방법으로 모델을 훈련시키는 것임
- 모델: 학습을 통해 만들어진(예측하는) 프로그램
- 모델은 새로운 입력에 대한 예측을 만듦  
<br><br>
- 지도학습은 기존의 데이터를 통해 모델을 학습시키고 학습시킨 모델로 새로운 입력에 대한 예측을 할 수 있으므로 내일의 날씨를 예측하거나 스팸 이메일을 분류하는 등의 일을 해결할 때 많이 사용함
- 지도학습에서 가장 어려운 일은 모델을 훈련시키는 것이 아니라 훈련 데이터를 만드는 것임
- 잘못된 입력과 타깃을 훈련 데이터에 포함시키면 잘못된 모델이 만들어질 수도 있고, 데이터의 수가 너무 적으면 모델을 충분히 훈련시킬 수 없기 때문
- 실전에서는 모델을 훈련시키기 위해 많은 시간이 훈련 데이터 만드는데에 소요됨

### 비지도 학습

- 비지도 학습은 타깃이 없는 훈련 데이터를 사용함
- 기업이 고객의 소비 성향에 따라 그룹을 지정하는 상황을 생각하면 이해하기 쉬움
- 그룹을 만들기 전까지는 어떤 그룹이 존재하는지, 또 몇개의 그룹이 만들어질지 알 수 없음
- 즉, 타깃이 없음
- 비지도 학습의 대표적인 예: clustering
- 모양이 비슷한 것으로 모아 그룹을 만들고, 그룹에서 새로운 모양이 발견되면 새로운 그룹이 생기는 형식
- 비지도 학습은 훈련 데이터에 타깃이 없으므로 모델의 훈련 결과를 평가하기 어렵다는 특징이 있음

### 강화학습

- 강화 학습은 주어진 환경으로부터 피드백을 받아 훈련함
- 머신 러닝 알고리즘 중 하나로 '에이전트'라는 것을 훈련시킴
- 훈련된 에이전트는 특정 환경에 최적화된 행동을 수행하고, 수행에 대한 '보상'과 '현재 상태'를 받음
- 에이전트의 목표는 '최대한 많은 보상을 받는 것'임
- 그래서 에이전트는 주어진 환경에서 아주 많은 행동을 수행하여 학습됨
- 대표적인 알고리즘으로는 Q-러닝(Q-learning), SARSA, 인공신경망을 사용한 DQN(Deep Q Network)등이 있음
- 강화 학습의 예로는 딥마인드의 알파고와 같은 게임이나 온라인 광고 등이 있음

### 규칙: 가중치와 절편

- 머신러닝은 스스로 '규칙'을 찾음
- 규칙이란 정확히 무엇일까?
- 대부분의 머신러닝 알고리즘은 훈련 데이터와 규칙의 관계를 식으로 표현할 수 있음
- 위의 예시에 적용해보면 (비 예시)
> 1.5(가중치) * x(입력) + 0.1(절편) = y(타깃) (y가 1 이상이면 다음날 비가 온다고 예측)
- 입력과 곱하는 수: 가중치
- 이와 더하는 수: 절편
- 여기서는 가중치와 절편이 각각 1개인 간단한 식이지만 딥러닝은 이것보다 훨씬 많은 개수(10만, 100만)의 가중치와 절편을 가짐

### 모델

- 모델은 머신러닝의 수학적 표현임
- 위의 수학식이 바로 모델임
- 엄밀히 정의하면 모델은 '훈련데이터로 학습된 머신러닝 알고리즘'을 말함
- 그리고 가중치와 절편을 합쳐 모델 파라미터(model parameter)라고 부름

### 손실 함수

- 만약 훈련 데이터 중 모델에 맞지 않는 예외 경우가 나오면 모델의 규칙(가중치, 절편)을 수정해야 함
- 이 때 모델의 규칙을 수정하는 기준이 되는 함수: 손실 함수(loss function)
- 예를 들어, 모델이 예측한 값과 타깃값의 차이를 계산하는 함수를 손실 함수로 정의할 수 있음

### 최적화 알고리즘으로 손실함수의 최솟값을 찾음
- 손실 함수는 단순히 예측한 값과 타깃의 차이를 정의하는 것이므로 손실 함수의 최솟값을 효율적으로 찾는 방법도 필요함
- 이 때 최솟값을 효율적으로 찾는 방법을 최적화 알고리즘이라고 부름

# 딥러닝

- 머신러닝 알고리즘에는 선형 회귀나 로지스틱 회귀와 같은 알고리즘들이 있음
- 인공신경망도 머신러닝 알고리즘 중 하나임
- 복잡한 문제를 해결하기 위해 인공신경망을 다양하게 쌓은 것을 딥러닝이라고 부름

### 딥러닝은 인공신경망으로 구성됨

- '딥러닝'이라는 용어는 인공신경망을 여러 겹으로 쌓은 모습에서 유래되었음
- 입력을 받아 출력으로 전달하는 인공신경망을 보면 뇌의 뉴런과 비슷한 점이 많음
- 뉴런은 자극을 받아들여 다음 뉴런으로 자극을 전달하는 역할을 수행하기 때문
- 인공신경망을 여러 겹 쌓은 것이 딥러닝이므로 보통 인공신경망과 딥러닝이라는 용어는 구분하지 않고 사용하는 경우가 많음
- 인공신경망은 뇌의 뉴런에서 영감을 받아 만든 것이긴 하지만, 실제 뇌가 작동하는 방식과는 많이 다름

## 딥러닝은 머신러닝이 처리하기 어려운 데이터를 더 잘 처리함

- 딥러닝과 머신러닝의 구체적인 차이점들 중 하나는 '처리하는 데이터'임
- 딥러닝에 잘 맞는 데이터: 이미지/영상, 음성/소리, 텍스트/번역 등의 비정형 데이터
- 머신러닝에 잘 맞는 데이터: 데이터베이스, 레코드 파일, 엑셀/csv 등에 담긴 정형 데이터
<br><br>
- 위와 같은 이유로, 딥러닝은 보통 '인지'와 관련된 문제를 잘 해결한다고 알려져 있음
- 예를 들어, 딥러닝은 이미지, 음성, 텍스트에 관련된 분야의 문제를 잘 해결함
- 반면에, 머신러닝은 암 환자의 검진 데이터나 회사의 매출 데이터와 같은 정형화된 데이터에 대한 문제를 잘 해결함
