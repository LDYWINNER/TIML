# NumPy

* NumPy는 파이썬의 핵심 과학 패키지 중 하나임
* 많은 머신러닝과 딥러닝 패키지가 넘파이를 기반으로 구현되었고 넘파이 인터페이스를 계승하고 있기 때문에 넘파이는 꼭 익혀야하는 파이썬 패키지 중 하나임


### 파이썬 리스트 복습

* 파이썬 리스트는 넘파이의 배열과 비슷함
* 2차원 배열: 리스트 안에 리스트가 있는 형태 - 바둑판 모양

예)

```python
my_list_2 = [[10, 20, 30], [40, 50, 60]]
```

* 위의 리스트의 크기는 보통 '2 x 3'의 크기라고 말함
* 즉, 리스트 안의 각 리스트가 하나의 행을 표시함


```python
print(my_list_2[1][1])
```

첫 번째 인덱스는 행의 인덱스, 두 번쨰는 열의 인덱스를 의미함

### 넘파이 준비하기

* 다차원 배열은 파이썬 리스트로 만들어 사용해도 되지만 파이썬 리스트로 만든 배열은 배열의 크기가 커질수록 성능이 떨어진다는 단점이 있음
* 넘파이는 저수준 언어로 다차원 배열을 구현했기 때문에 배열의 크기가 커져도 높은 성능을 보장함
* 또한 배열을 이용한 다양한 통계, 수학 함수도 제공함

1. 코랩에서 넘파이 임포트하고 버전 확인

```python
import numpy as np
print(np.__version__)
```

### 넘파이로 배열 만들기

* 넘파이는 파이썬의 리스트처럼 숫자와 문자열을 함께 담을 수 없음

1. array() 함수로 2차원 배열 만들기


```python
my_arr = np.array([[10, 20, 30], [40, 50, 60]])
print(my_arr)
```

2. type() 함수로 넘파이 배열인지 확인하기

* array() 함수는 파이썬 리스트를 입력 받아 넘파일 배열을 만듦

```python
type(my_arr)
```

* 넘파이 배열 클래스: ndarray
* 코랩의 마지막줄은 print() 함수를 사용하지 않아도 결과가 출력됨

3. 넘파이 배열에서 요소 선택하기
* 넘파이 배열의 요소를 선택하는 방법은 파이썬 리스트에서 요소를 선택하는 방법과 동일


```python
my_arr[0][2]
```

4. 넘파이 내장 함수 사용하기

* 파이썬 리스트의 요소를 모두 더하려면 for문을 사용해야 함
* 넘파이는 배열에 대한 여러 수학 함수를 제공하기 때문에 반복문을 쓰지 않고도 배열의 합을 계산할 수 있음
* sum() 함수는 배열의 요소를 모두 더해주는 함수임
* for 문을 사용하는 것보다 실행속도가 매우 빠르고 코드가 간결함

```python
np.sum(my_arr)
```

### 맥플롯립으로 그래프 그리기

* Matplotlib은 파이썬 과학 생태계의 표준 그래프 패키지임
* 다른 그래프 도구들도 괜찮지만 맥플롯립에서 제공하는 기능으로 대부분의 그래프를 그릴 수 있음

```python
import matplotlib.pyplot as plt
```

* 선 그래프와 산점도 자주 사용
* 두 그래프는 x축을 기준으로 y축의 변화추이를 살펴보기 편리해 데이터를 분석할 때 자주 사용함

1. 선 그래프
* 선 그래프를 그리려면 x축의 값과 y축의 값을 맷플롯립의 plot() 함수에 전달해야 함
* 그런 다음 show() 함수를 이용하여 화면에 출력하면 됨

예) y = x ^ 2

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

2. 산점도 scatter plot

* 산점도는 데이터의 x축, y축 값을 이용하여 점으로 그래프를 그린 것임
* show() 함수를 사용하지 않아도 맷플롯립이 자동으로 그래프를 그려주지만 이용하면 조금 더 깔끔하고 보기좋은 그래프가 그려짐

```python
plt.scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.show()
```

3. 넘파이 배열로 산점도 그리기

* 파이썬의 과학 패키지들은 예외 없이 넘파이 배열로 데이터를 주고받을 수 있도록 만들어져 있음
* 넘파이의 random.randn() 함수를 사용하여 표준정규분포를 따르는 난수(random number)를 만들고 그 값을 이용하여 산점도 그려보기

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.scatter(x, y)
plt.show()
```

* 데이터 분석과 그래프 도구는 뗄래야 뗄 수 없는 사이임
* 위와 같은 방법으로 데이터를 시각화하면 데이터에서 직관을 얻기 쉬움
* 딥러닝에서도 데이터 시각화는 필수
